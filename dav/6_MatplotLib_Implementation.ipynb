# Matplotlib Implementation
### Scatter Plot, Box Plot, Voiline Plot, Bubble Plot, Area Chart

# Scatter Plot:

Purpose: Visualizes the relationship between two continuous variables. Each data point represents a pair of values (x, y).
Matplotlib elements:
plt.scatter(x, y): Creates a scatter plot with x-axis values in x and y-axis values in y.
Customization options include markers, colors, size, labels, and title using functions like plt.xlabel(), plt.ylabel(), etc.
Useful for identifying trends, correlations, and outliers.

# Box Plot:

Purpose: Summarizes the distribution of a continuous variable across a categorical variable (groups).
Matplotlib elements:
plt.boxplot(data, notch=True, vert=True, patch_artist=True): Creates a box plot with data for each group in data.
The box represents the interquartile range (IQR) with the median as a line. Whiskers extend to data points within 1.5 IQRs from the quartiles. Outliers are plotted as individual points.
Useful for comparing distributions across groups and identifying potential differences.

# Violin Plot:

Purpose: Similar to box plots, violin plots depict the distribution of a continuous variable across categories. Offers additional information about the data density.
Matplotlib elements:
Specialized libraries like seaborn are often used for violin plots.
The violin shape shows the kernel density estimation of the data, with wider areas indicating higher density.
Offers more insights into the distribution compared to box plots.

# Bubble Plot:

Purpose: Similar to scatter plots, bubble plots represent relationships between two continuous variables. A third quantitative variable is encoded by the size of the bubbles.
Matplotlib elements:
plt.scatter(x, y, s=size_data): Creates a scatter plot with bubble sizes determined by size_data.
Useful for visualizing the relationship between three variables, where size adds another dimension to the analysis.

# Area Chart:

Purpose: Depicts the cumulative sum or trend of a continuous variable over time or another continuous variable. Often used to represent stacked data.
Matplotlib elements:
plt.fill_between(x, y1, y2, where=condition): Creates an area chart with x-axis values in x, data for the bottom boundary (y1), data for the top boundary (y2), and a condition to define the filled area.
Useful for visualizing trends over time, proportions of a whole when stacked, or confidence intervals when shaded.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generating some random data for demonstration
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)
sizes = np.random.randint(10, 100, 100)
colors = np.random.rand(100)
```

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Box Plot
data = np.random.randn(100, 3)  # Three sets of random data
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(8, 6))
plt.violinplot(data, showmedians=True)
plt.title('Violin Plot')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.show()
```

```python
# Bubble Plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.title('Bubble Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Area Chart
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 6))
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.5)
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='red')
plt.title('Area Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
