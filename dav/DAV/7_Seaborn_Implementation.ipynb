# Seaborn Implementation
### Line graph, bar graph, histograph, scatter plot, box plot

# Theory:

## Line Graph (Lineplot):

Similar to Matplotlib's line plots, Seaborn's sns.lineplot() helps visualize trends or relationships between continuous variables over time or categories.
It offers a more concise and aesthetically pleasing syntax compared to Matplotlib.

## Bar Graph (Barplot):

Seaborn's sns.barplot() creates bar charts for categorical comparisons.
It inherits the core functionality of Matplotlib's bar charts but provides a more streamlined approach.
Seaborn allows for stacking bars to compare multiple categories on top of each other, dodging bars to reduce overlap in case of many categories, and hue mapping to represent another categorical variable using color.

## Histogram (Distplot):

Seaborn's sns.distplot() simplifies creating histograms for visualizing the distribution of continuous data.
It builds upon Matplotlib's plt.hist() function but offers options for kernel density estimation (smoother curve) and rug plots (data points along the axis).
Seaborn also facilitates customizing the number of bins, color, and adding statistical information like quartiles.

## Scatter Plot:

Seaborn's sns.scatterplot() provides a foundation for creating scatter plots similar to Matplotlib.
It offers a more concise syntax and leverages additional features like hue mapping to represent a third categorical variable using color or size.
Seaborn also allows for regression line fitting and customization of marker style and size.

## Box Plot (Boxplot):

Seaborn's sns.boxplot() simplifies creating box plots to compare distributions across categorical variables.
It inherits the core functionality of Matplotlib's box plots but offers a more visually appealing design and easier customization.
Seaborn allows for hue mapping to represent another categorical variable using box color or faceting the plot to create multiple box plots for different categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generating some random data for demonstration
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Line Graph
plt.figure(figsize=(8, 6))
sns.lineplot(x=x, y=y)
plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Bar Graph
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 35, 30, 25, 40]

plt.figure(figsize=(8, 6))
sns.barplot(x=categories, y=values)
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python

# Histogram
data = np.random.randn(1000)

plt.figure(figsize=(8, 6))
sns.histplot(data, bins=30, kde=True)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Generating some random data for demonstration
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x, y=y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python

# Box Plot
data = np.random.randn(100, 3)  # Three sets of random data

plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title('Box Plot')
plt.xlabel('Data Set')
plt.ylabel('Value')
plt.show()
```
