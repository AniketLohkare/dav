# Matplotlib Implementation
### Line Chart, Bar Chart, Histogram, Pie Chart



# LINE CHART THEORY
Line Charts:

Purpose: Line charts depict trends or relationships between continuous data points over time or along a category.
Matplotlib elements:
plt.plot(x, y): Creates a line plot with x-axis values in x and y-axis values in y.
Customization options include markers, line styles, colors, labels, and titles using functions like plt.xlabel(), plt.ylabel(), plt.title(), etc.

# Bar Chart Theory
Bar Charts:

Purpose: Bar charts represent categorical data comparisons. Each category has a bar proportional to the value it represents.
Matplotlib elements:
plt.bar(x, height): Creates a bar chart with categories on the x-axis (x) and heights of bars (height).
Customization options include bar width, color, labels, and title using similar functions as line charts.
Stacked bar charts can be created using plt.bar_stack to visualize multiple categories on top of each other.

# Histogram Theory
Histograms:

Purpose: Histograms represent the distribution of continuous data. They show the frequency of data points falling within specific ranges (bins).
Matplotlib elements:
plt.hist(data): Creates a histogram for the data in the provided list (data).
Customization options include the number of bins, color, and labels using functions like plt.xlabel(), plt.ylabel(), etc.

# Pie Chart Theory
Pie Charts:

Purpose: Pie charts represent proportions of a whole using circular slices. Each slice corresponds to a category and its size reflects its percentage of the total.
Matplotlib elements:
plt.pie(data, labels=labels): Creates a pie chart with data values (data) and corresponding category labels (labels).
Customization options include colors, wedge labels (percentages), and a title using functions like plt.title().

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generating some random data for demonstration
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Line Chart


plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
# Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 35, 30, 25, 40]

plt.figure(figsize=(8, 6))
plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie Chart
sizes = [30, 20, 25, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
